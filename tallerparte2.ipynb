```python
#ejercicio 10
c= input('introdusca una cadena: ')
c1=''
for n in range((len(c))):
    v=ord(c[n]) #cada letra toma su numero de python
    if v>=97 and v<=122:# son los ord de a a z en minuscula
        v=v-32 #las mayusculas estan 32 puestos antes que las minusculas
    elif v==255: #á
        v=193 #Á
    elif v==237: #í
        v=205 #Í
    elif v==243: #ó
        v=211 #Ó
    elif v==233: #é
        v=201
    elif v==250: #ú
        v=218
    elif v==241: #ñ
        v=209
    elif v==252: #ü
        v=220
    c1 += chr(v)

print (c1)
```

```python
#ejercicio17
lista=[] #Lista vacia para el almacenamiento de datos
pibote=[] #almacenamiento de los pibotes encontrados (numeros mayores 
          #hasta el momento)
cantidad=int(input('ingrese la cantidad de numeros de la lista: '))#Cantidad de 
                                                            #datos a recolectar
for i in range(cantidad): #ciclo para recoleccion de valores de la lista
    lista.append(float(input(f'Escriba el numero {i+1}\
    de la lista ordenada: ')))
parar=lista[0] #se toma el primer elemento de la lista como pibote y se 
               #genera la condicion parar

for i in range(len(lista)):#Ciclo para recorrer la lista ingresada
    if lista[i]>parar:     #compara el pibote con el numero i de la lista
        parar=lista[i]     # si el valor de la lista es mayor al pibote actual
                           #lo sustituye hasta acabar de recorrer toda la lista
        pibote.append(parar) #Se almacena el pibote actual
print(f'El segundo mayor numero de la lista es {pibote[len(pibote)-2]} \
y se encuetra en la posicion numero {lista.index(pibote[len(pibote)-2])+1} \
de la lista') #Se toma el penultimo pibote guardado
```

```python
#ejercicio 18

matriz=[] #crea la variable matriz
filas=int(input('ingrese el numeor de filas: ')) #Numero de filas de la matriz
columnas=int(input('ingrese el numeor de columnas: ')) #Numero de columnas 

while filas != columnas: #Como requisito se debe ingresar una matriz cuadrada
    print('ERROR infgrese una matriz cuadrada')
    filas=int(input('ingrese el numeor de filas: ')) #Se ingresa nuevaente las 
                                     #dimensiones ahsta que cumpla la condicion
    columnas=int(input('ingrese el numeor de columnas: '))
   
for i in range(filas): 
    matriz.append([0]*columnas) #genera una matriz de 0 con dimnesiones 
                                #ingresadas anteriormente 

for i in range(filas): #Ciclo para la recoleccion de datos de la matriz 
    for j in range(columnas):
        matriz[i][j]=int(input(f'ingrese el elmento {i+1},{j+1} de la \
matriz: '))
print('La matriz original es ', matriz)
U=matriz #Se copia la matriz a una nueva variable U que se editara

#Recorrido por la matriz y sustitucion de 0
for i in range(filas): #Ciclo para el recorrido por la matriz
    for j in range(columnas): 
        if i>j: #Para una matriz superior se cumple que para la posicion i>j 
                #aparece un cero
            U[i][j]=0 #Si se cumple la condicion se coloca un cero en dicha 
                      #posicion de la matriz U
print('La matriz diagonal superior U es: ', U)

#Busqueda de 1 en las filas
contador=[0]*filas #Variable para almacenar los numero 1 contados de cada fila
for i in range(filas): #Ciclo para el recorrido de la matriz
    for j in range(columnas): 
        if U[i][j]//1 ==1: #Si en la posicion se encuentra un 1 se sumara una 
                           #unidad al contador de dicha fila
            contador[i]=contador[i]+1

parar=contador[0] #se toma el primer elemento de la variable contar 1 como 
                  #pibote en la busqeuda del mayor numero

for i in range(len(contador)):
    if contador[i]>parar:  #compara el pibote con el numero i de la lista 
                           #"contador" (donde esta almacenada la cantidad de 
                           #unos encontrada en cada fila)
        parar=contador[i]  # si es mayor al pibote actual lo sustituye hasta 
                           #acabar de recorrer toda la lista y asi encontrar 
                           #el mayor numero presente en la lista
print(f'El mayor numero de unos en una fila de U es de {parar} y se encuetra\
      en la fila numero {contador.index(parar)+1} de la matriz: ')
```

```python
#ejercicio 19

cant=(int(input('Digite la cantidad de elemntos desea en la lista: ')))
lista=[]
#llenando lista
for i in range(cant):
    lista.append((input('Ingrese el elemento numero '+str(i+1)+': ' )))
A=[]    #matriz donde se guarda cada sublista de numeros consecutivos
tamaño=[] #vector con los tamaños de cada sublista
for i in range(len(lista)-1):
    A.append([]) #se crea una sublista
    A[i].append(lista[i])#le agrega el primer elemento analizado
    for j in range(i+1,len(lista)):
        if lista[j-1]<=lista[j]:
            A[i].append(lista[j])
        else:
            break
    tamaño.append(len(A[i])) #introduce el tamaño de cada lista
cantMayor=tamaño.count(max(tamaño))
pos=tamaño.index(max(tamaño))#posicion de la primera sublista mas larga
sublista=[]
if(cantMayor==1):
    print('La sublista mas larga ordenada ascendentemente es: ',A[pos])
else:
    a=-1 #es un valor que dice la posicion desde donde se analizara
    for i in range(cantMayor):
        a=tamaño.index(max(tamaño),a+1)
        sublista.append(A[a])
    print('las sublistas mas largas ordenadas ascendentemente son: ',sublista)
```

```python
#ejercicio 20
#se generan las variables en el programa
matriz_orig=[]# variable de la matriz original
matriz_rot=[]# variable para la matriz rotada
#se generan las variables  de filas y columnas para la matriz
filas=(int(input('Ingrese el numero de filas de la matriz: ')))
columnas=(int(input('Ingrese el numero de columnas de la matriz: ')))
for i in range(filas):#iteracion de la variable filas con respecto a su rango
    matriz_orig.append([])#a la matriz con respecto al rango ingresado
    for j in range (columnas):
        matriz_orig[i].append(int(input(f'Escriba el numero en {j+1}\
                   de la fila {i+1}: ')))
for i in range(columnas):#iteracion de la variable columnas con el rango
        matriz_rot.append([])
        for j in range(filas):#iteracion  de filas con el valor ya dado
            
            matriz_rot[i].append(matriz_orig[j][columnas-1-i])#relacion 
            #matematica entre las funciones i y j
print()
print('La matriz original ingresada es la siguiente: ')
for i in range(filas):
    print(matriz_orig[i], end=' ')
    print()
print()
print('la matriz rotada a 90 grados, con datos de la matriz original es: ')
for i in range(columnas):
    print(matriz_rot[i], end=' ')
    print()
```
