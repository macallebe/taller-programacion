```python
#ejercicio 6
def suma_digitos(x):# se define una funcion del residuo
    cont=0
    while x>0:
        residuo=x%10
        x=x//10
        if residuo>0:
            cont=cont+residuo
    return cont
def numero_fibonacci(x):# se define una nueva funcion
    num1=0
    num2=1
    resultado= False
    while(num1<=x):
        if (num1==x):
            resultado= True
        suma=num1+num2
        num1=num2
        num2=suma
    return resultado

while True:#ciclo cuando la condicion es verdadera permite ingresar un numero
    #entero y corrige si es negativo
    try:
        num=int(input('escriba un número entero = '))
        break
    except ValueError:
        print('el valor ingresado es incorrecto. ')
        
cont=suma_digitos(num)
if cont<0:
    cont=cont*(-1)
    
resultado=numero_fibonacci(cont)
if resultado:
    print('la suma de los digitos en valor absoluto de numero ingresado \
          es',cont,'.\nY el numero',cont,'es dela serie fibonacci.')
else:
    print('la suma de los digitos en valor absoluto es el numero',cont,'\n\
          Y no es de la serie fibonacci.')
```

```python
#ejercicio #11
conv=0 #contador de vocales mayúsculas
cont=0 #contador de tilde
cond=0 #contador de digitos
cone=0 #contador de espacios
conr=0 #contador de palabras reservadas por python

rese=['if','and','del','assert','in','else','return','break','global',\
      'exec','import','yield','def','elif','except','continue','not','try',\
      'or','while','print','assert','for','is','from','class','finally',\
      'lambda','return','raise']
# se creó una lista con las palabras reservadas de python 
t= input('Digite su texto: ')

for i in range(0,len(t)):#se define un rango para que me cuente letra por letra
#permite definir e identificar si es mayuscula y si es vocal
    if t[i].isupper()==True:
        if ord(t[i])==65 or ord(t[i])==69 or ord(t[i])==73 or ord(t[i])==79\
        or ord(t[i])==85 or ord(t[i])==193 or ord(t[i])==201 or ord(t[i])==205\
        or ord(t[i])==211 or ord(t[i])==218:
           
            conv+=1 #permite agregar al contador de vocales mayusculas
    #permite identificar si hay letras con tilde
    if ord(t[i])==225 or ord(t[i])==233 or ord(t[i])==237 or ord(t[i])==243\
    or ord(t[i])==250 or ord(t[i])==193 or ord(t[i])==201 or ord(t[i])==205\
    or ord(t[i])==211 or ord(t[i])==218:
           
        cont+=1 #sumar al contador de letras con tilde
        
    if t[i].isdigit()==True: #se difita una secuencia para comprobar si hay
        #digitos
         
         cond+= 1 #Adiciona al contador de digitos
    if ord(t[i])==32:# permite cuantificar la cantidad de espacios
         cone+=1#Adiciona al contador de espacios
         
d= t.split() #se lleva la cadena digitada de texto a una lista para evaluar
#cada elemento
for s in range(len(d)):# se itera en la longitud de la lista creada
    if d[s] in rese:#se buscan las palabras reservadas
        
        conr+= 1# se suma al contador de palabras reservadas
        
print('se ingresaron',conv,'Vocales Mayusculas')
print('se ingresaron',cont,'Letras con tilde')
print('se ingresaron',cond,'Digitos')
print('se ingresaron',cone,'Espacios')
print('se ingresaron',conr,'palabras reservadas')
```

```python
#ejercico 16
#digitar el numero de caracteres de la lista
long = int(input('digite la longitud de la lista: '))
lista = []
for i in range(long):   #ciclo para agregar cada termino de la lista
    lista.append(float(input('inserte un numero: ')))

lis_orden = []  #defino una nueva lista ordenada
lis_orden = sorted(lista)   #la lista sera la ingresada ordenada ascendente
          
if lis_orden == lista:  #condicional de que la lista ingresada sea ordenada
    print('la lista esta ordenada')

    #ingresar el numero que se desea añadir
    numero = float(input('inserte un numero para agregar a la lista: '))    
    lista.append(numero)    #añadirlo a la lista
    lis_orden = sorted(lista)   #organizar la nueva lista ascendentemente
    posicion = lis_orden.index(numero)  #dar la posicion del numero ingresado
    print('la posicion indexada del numero inresado es: ',posicion)

else:
    print('la lista no esta ordenada')
```
